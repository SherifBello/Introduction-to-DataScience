```python
# Reading the dataset of International football results from 1872 and 2026
import pandas as pd
import matplotlib.pyplot as plt
football = pd.read_csv('../../../archive/results.csv')
football.head()
```

```python
football.info()
```

```python
# Dataset Shape
football.shape
```

```python
# plot showing countries amd frequency of matches played in the dataset
country_count = football.value_counts(football['home_team'].values, sort=True)
plt.rcParams['figure.figsize'] = [6, 50]
country_count.plot.barh()
```

```python
# Plot showing the types of tournaments played in the dataset
tournament_type = football.value_counts(football['tournament'].values, sort=True)
plt.rcParams['figure.figsize'] = [6, 30]
tournament_type.plot.barh()
```

```python
# Plot showing the home and away scores of matches played in the dataset
Home_Score = football['home_score']
Away_Score = football['away_score']
Country = football['country']

plt.barh(Country, Home_Score)
plt.barh(Country, Away_Score)
plt.rcParams['figure.figsize'] = [6, 30]
plt.show()
```

```python
football.plot(x='country', y=['home_score', 'away_score'],
        kind='bar',
        stacked=True,
        title='International Football Scores by Country',
        xlabel='Country',
        ylabel='Score')
```

```python
Goals_Scored = football[(football['home_score'] > 0) & (football['home_score'] < 10)]      
Goals_Scored['home_score'].plot(kind = 'hist',bins = 9,figsize = (6,6),title = 'Distribution of Goals Scored by Home Team')
plt.show() 
```

```python
Goals_Scored = football[(football['away_score'] > 0) & (football['away_score'] < 10)]      
Goals_Scored['away_score'].plot(kind = 'hist',bins = 9,figsize = (6,6),title = 'Distribution of Goals Scored by Away Team')
plt.show() 
```
