```python
# 56. Generate a generic 2D Gaussian-like array

def gaussian_2d(shape=(100, 100), center=None, sigma=(10, 10)):
    """
    shape: (rows, cols) = (y, x)
    center: (μx, μy). Default = center of array
    sigma: (σx, σy) std dev in x and y
    """
    h, w = shape
    if center is None:
        center = (w/2, h/2) # x, y
    
    x = np.arange(w)
    y = np.arange(h)
    X, Y = np.meshgrid(x, y)
    
    cx, cy = center
    sx, sy = sigma
    
    g = np.exp(-((X-cx)**2/(2*sx**2) + (Y-cy)**2/(2*sy**2)))
    return g

g = gaussian_2d(shape=(200, 200), center=(100, 80), sigma=(20, 40))
print(g.shape)
print(g.max()) 
```

```python
# 62. Considering two arrays with shape (1,3) and (3,1), how to compute their sum using an iterator?

A = np.array([[1, 2, 3]]) # shape (1,3)
B = np.array([[4], [5], [6]]) # shape (3,1)

result = np.zeros((3,3))

it = np.nditer([A, B, result],
               flags=['external_loop', 'buffered'],
               op_flags=[['readonly'], ['readonly'], ['writeonly']])

for a, b, c in it:
    c[...] = a + b # broadcasting happens inside

print(result)
```

```python
# 63. Create an array class that has a name attribute 

class NamedArray(np.ndarray):
    def __new__(cls, input_array, name=""):
        # Convert input to ndarray, then view as NamedArray
        obj = np.asarray(input_array).view(cls)
        obj.name = name
        return obj
    
    def __array_finalize__(self, obj):
        # Called when new arrays are created from this one
        if obj is None: return
        self.name = getattr(obj, 'name', "")

# Usage
a = NamedArray([[1, 2, 3], [4, 5, 6]], name="my_matrix")
print(a.name)
print(a.shape) 
print(a + 10)
```

```python
# 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])?

P0 = np.array([[0,0], [1,1], [2,0]]) # shape (n, 2)
P1 = np.array([[1,0], [2,2], [3,0]]) # shape (n, 2)
p  = np.array([0.5, 0.5])            # shape (2,)

# Vector from P0 to P1 for each line
v = P1 - P0  # (n, 2)

# Vector from P0 to p
w = p - P0   # (n, 2)

# Cross product magnitude in 2D: |v_x * w_y - v_y * w_x|
cross = np.abs(v[:,0]*w[:,1] - v[:,1]*w[:,0])

# Line length
norm_v = np.linalg.norm(v, axis=1)

# Distance = |cross| / |v|
dist = cross / norm_v
print(dist)
```

```python
# 79. Consider 2 sets of points P0,P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P[j]) to each line i (P0[i],P1[i])?

P0 = np.array([[0,0], [1,1]])     # shape (n_lines, 2)
P1 = np.array([[1,0], [2,2]])     # shape (n_lines, 2)
P  = np.array([[0.5,0.5], [3,1]]) # shape (n_points, 2)

# Add axes for broadcasting: (n_lines,1,2) vs (1,n_points,2) vs (n_points,1,2)
v = P1[:, None, :] - P0[:, None, :]  # (n_lines, 1, 2) line vectors
w = P[None, :, :] - P0[:, None, :]   # (n_lines, n_points, 2) vec from P0 to each P

# 2D cross product |v x w| for all pairs
cross = np.abs(v[...,0]*w[...,1] - v[...,1]*w[...,0])  # (n_lines, n_points)

# Line lengths, keep dim for broadcast
norm_v = np.linalg.norm(v, axis=2)  # (n_lines, 1)

# Distance matrix: rows=lines, cols=points
dist = cross / norm_v  # (n_lines, n_points)
dist = dist.T          # (n_points, n_lines) if you want points as rows
```

```python
# 80. Consider an arbitrary array, write a function that extract a subpart with a fixed shape and centered on a given element (pad with a fill value when necessary)

def extract_centered(arr, center, shape, fill=0):
    """
    arr: n-dim array
    center: tuple of indices, e.g. (i,j) for 2D
    shape: tuple of output shape, e.g. (5,5)
    fill: value used for padding
    """
    arr = np.asarray(arr)
    center = np.asarray(center)
    shape = np.asarray(shape)

    # Half window on each side. For even shapes, bias right/bottom
    half = shape // 2
    extra = shape % 2 # handles even sizes

    # Amount to pad: [before, after] for each axis
    pad_width = [(h, h + e) for h, e in zip(half, extra)]
    arr_padded = np.pad(arr, pad_width, mode='constant', constant_values=fill)

    # Adjust center because of padding
    center_pad = center + half

    # Build slices for each dim
    slices = tuple(slice(c - h, c + h + e)
                   for c, h, e in zip(center_pad, half, extra))

    return arr_padded[slices]

A = np.arange(25).reshape(5,5)
print (A)
# center at (2,2), 3x3 window
print('\n',(extract_centered(A,(2,2),(3,3))),'\n')
# Center at corner (0,0), 3x3 window, pad with -1
print(extract_centered(A,(0,0),(3,3),fill=-1))
```

```python
# 85. Create a 2D array subclass such that Z[i,j]==Z[j,i]
import numpy as np
class SymmetricArray(np.ndarray):
    def __new__(cls, input_array):
        obj = np.asarray(input_array).view(cls)
        i_upper = np.triu_indices_from(obj, k=1)
        obj[i_upper] = obj.T[i_upper]
        return obj
    
    def __setitem__(self, index, value):
        super().__setitem__(index, value)
        if isinstance(index, tuple) and len(index) == 2:
            i, j = index
        # Ensure symmetry: Z[i,j] = Z[j,i]
        super().__setitem__((j, i), value)

# usage
Z = SymmetricArray(np.zeros((4,4)))
Z[0, 3] = 5
print(Z)
print(Z[0,3]==Z[3,0])
```

```python
# 88. How to implement the Game of Life using numpy arrays? (★★★)
def game_of_life_step(board):
    neighbors = sum(np.roll(np.roll(board, i, 0), j, 1)
                    for i in (-1, 0, 1) for j in (-1, 0, 1)
                    if (i != 0 or j != 0))
    return (neighbors == 3) | (board & (neighbors == 2))
# Example usage
board = np.random.choice([0, 1], size=(5, 5))
print("Initial board:\n", board)
new_board = game_of_life_step(board)
print("New board:\n", new_board)
```

```python
# 99. Given an integer n and a 2D array X, select from X the rows which can be intepereted as draws fron a multinomial distributon with n degrees, i.e., the rows which only contain integers and which sum to n.
def multinomial_rows(X, n):
    """ Return rows of X that are valid multinomial draws with n trials """
    X = np.asarray(X)
    # Check if all elements are integers and rows sum to n
    is_integer = np.all(np.mod(X, 1) == 0, axis=1)
    sums_to_n = np.sum(X, axis=1) == n
    valid_rows = X[is_integer & sums_to_n]
    return valid_rows
# Example usage
X = np.array([[2, 1, 0], 
              [1, 2, 2], 
              [0, 0, 3]])
n = 3
result = multinomial_rows(X, n)
print(result)
```

```python
# 100. Compute bootsrapped 95% confidence intervals for the mean of a 1D array X (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means).
def bootstrap_confidence_interval(X, N=1000, alpha=0.05):
    """ Compute bootstrap confidence interval for mean of X """
    X = np.asarray(X)
    means = np.empty(N)
    for i in range(N):
        sample = np.random.choice(X, size=len(X), replace=True)
        means[i] = np.mean(sample)
    lower = np.percentile(means, 100 * (alpha / 2))
    upper = np.percentile(means, 100 * (1 - alpha / 2))
    return lower, upper
# Example usage
X = np.random.rand(100)  # Sample data
lower, upper = bootstrap_confidence_interval(X)
print(f"95% confidence interval for the mean: [{lower:.2f}, {upper:.2f}]")
```
