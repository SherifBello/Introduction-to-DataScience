```python
# Importing the required libraries
# Reading the Birds dataset and displaying the first 5 rows
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../../data/birds.csv')
birds.head()
```

```python
# Filtering the birds dataset to get only the Snow goose entries
Snow_geese = birds [birds['Name'] == 'Snow goose']
Snow_geese
```

```python
# Filtering the birds dataset to get only the Ducks/Geese/Waterfowl category
category = birds[birds['Category'] == 'Ducks/Geese/Waterfowl']
category
```

```python
# ploting the maximum wingspan of Ducks/Geese/Waterfowl in Minnesota
plt.title('Max Wingspan of Ducks/Geese/Waterfowl in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Ducks/Geese/Waterfowl')
wingspan = category.MaxWingspan 
wingspan.plot()
```

```python
# Plotting the Max Wingspan of Ducks/Geese/Waterfowl with their names on the x-axis
plt.title('Max Wingspan of Ducks/Geese/Waterfowl in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.xlabel('Ducks/Geese/Waterfowl')
plt.xticks(rotation=45)
x = category['Name'] 
y = category['MaxWingspan']

plt.plot(x, y)

plt.show()
```

```python
# Showing the Max Wingspan of Ducks/Geese/Waterfowl with their names on the x-axis and highlighting the Snow goose
plt.title('Max Wingspan of Ducks/Geese/Waterfowl in Centimeters')
plt.ylabel('Wingspan (CM)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(category)):
    x = category['Name'][i]
    y = category['MaxWingspan'][i]
    plt.plot(x, y, 'bo')
    if category['Name'][i] == 'Snow goose':
        plt.text(x, y * (1 - 0.05), category['Name'][i], fontsize=12)
    
plt.show()
```

```python
# Showing the categories of Ducks/Geese/Waterfowl in Minnesota with their names on the x-axis using bar chart
category.plot(x='Name',
        kind='bar',
        stacked=True,
        title='Ducks/Geese/Waterfowl of Minnesota')
```

```python
# Genus counts of Ducks,Geese & Waterfowl in Minnesota
Genus_count = category.value_counts(category['Genus'].values, sort=True)
plt.rcParams['figure.figsize'] = [6, 12]
Genus_count.plot.barh()
```

```python
# Superimposing the minimum and maximum lengths of Ducks/Geese/Waterfowl with their Genus on the y-axis using barh chart
minLength = category['MinLength']
maxLength = category['MaxLength']
genus = category['Genus']

plt.barh(genus, maxLength)
plt.barh(genus, minLength)

plt.show()
```
