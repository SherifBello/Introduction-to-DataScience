```python
Sisters = ('Rahimat', 'Nafisat', 'Samirat', 'Zainab')
print('Sisters:', Sisters)
Brothers = ('Khalid', 'Hafeez', 'Kazeem')
print('Brothers:', Brothers)
siblings = Sisters + Brothers
print('Siblings:', siblings)
print('Number of siblings:', len(siblings))
siblings = list(siblings)
siblings
Father = 'Bello'
Mother = 'Fatima'
family_members = siblings + [Father] + [Mother]
print('Family members:', family_members)
```

```python
Ra, Na, Sa, Za, Kh, Ha, Ka, *parents = family_members
print('Siblings:', Ra, Na, Sa, Za, Kh, Ha, Ka)
print('Parents:', parents)
```

```python
fruits = ('banana', 'orange', 'mango', 'lemon')
vegetables = ('Tomato', 'Potato', 'Cabbage', 'Onion', 'Carrot')
animal_products = ('milk', 'meat', 'eggs', 'cheese')
food_stuff_tp = fruits + vegetables + animal_products
print('Food stuff tuple:', food_stuff_tp)
food_stuff_lt = list(food_stuff_tp)
print('Food stuff list:', food_stuff_lt)
print(food_stuff_lt[6])
print(food_stuff_lt[:3])
print(food_stuff_lt[-3:])
del food_stuff_tp
print(food_stuff_tp)
```
