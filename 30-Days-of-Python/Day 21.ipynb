### LEVEL 1

```python
class Statistics:
    """A class to compute various statistical measures of a list of numbers."""
    def __init__(self, data):
        self.data = data
        self.sorted_data = sorted(data)
        self.n = len(data)

    def count(self):
        """Return the count of numbers in the data."""
        return self.n

    def sum(self):
        """Return the sum of the numbers in the data."""
        return sum(self.data)

    def min(self):
        """Return the minimum value in the data."""
        return self.sorted_data[0]

    def max(self):
        """Return the maximum value in the data."""
        return self.sorted_data[-1]

    def range(self):
        """Calculate the range of the data."""
        return self.max() - self.min()

    def mean(self):
        """Calculate the mean of the data."""
        return sum(self.data) / self.n

    def median(self):
        """Calculate the median of the data."""
        sorted_data = self.sorted_data
        n = self.n
        mid = n // 2
        if n % 2 == 0:
            return (sorted_data[mid - 1] + sorted_data[mid]) / 2
        else:
            return sorted_data[mid]

    def mode(self):
        """Calculate the mode of the data."""
        from collections import Counter
        data_count = Counter(self.data)
        mode_data = data_count.most_common(1)
        return mode_data[0][0] if mode_data else None

    def variance(self):
        """Calculate the variance of the data."""
        mean = self.mean()
        return sum((x - mean) ** 2 for x in self.data) / self.n

    def std_dev(self):
        """Calculate the standard deviation of the data."""
        return self.variance() ** 0.5

    def percentile(self, percentile):
        """Calculate the given percentile of the data."""
        if not 0 <= percentile <= 100:
            raise ValueError("Percentile must be between 0 and 100.")
        index = int(percentile / 100 * self.n)
        return self.sorted_data[index]

    def frequency_distribution(self):
        """Calculate the frequency distribution of the data."""
        freq = {}
        for item in self.data:
            freq[item] = freq.get(item, 0) + 1
        return sorted(freq.items(), key=lambda x: x[1], reverse=True)

ages = [31, 26, 34, 37, 27, 26, 32, 32, 26, 27, 27, 24, 32, 33, 27, 25, 26, 38, 37, 31, 34, 24, 33, 29, 26]
age_stats = Statistics(ages)

print("Count:", age_stats.count())
print("Sum:", age_stats.sum())
print("Min:", age_stats.min())
print("Max:", age_stats.max())
print("Range:", age_stats.range())
print(f"Mean: {age_stats.mean()}")
print(f"Median: {age_stats.median()}")
print(f"Mode: {age_stats.mode()}")
print(f"Variance: {age_stats.variance():.2f}")
print(f"Standard Deviation: {age_stats.std_dev():.2f}")
print(f"Frequency Distribution: {age_stats.frequency_distribution()}")
```

### LEVEL 2

```python
class PersonAccount:
    def __init__(self, first_name, last_name, incomes=None, expenses=None, properties=None):
        self.first_name = first_name
        self.last_name = last_name
        self.incomes = incomes if incomes is not None else []
        self.expenses = expenses if expenses is not None else []
        self.properties = properties if properties is not None else []

    def total_income(self):
        return sum(self.incomes)

    def total_expenses(self):
        return sum(self.expenses)

    def account_info(self):
        return f"{self.first_name} {self.last_name} has a total income of {self.total_income()} and total expenses of {self.total_expenses()}."
    
    def add_income(self, amount):
        self.incomes.append(amount)

    def add_expense(self, amount):
        self.expenses.append(amount)

    def account_balance(self):
        return self.total_income() - self.total_expenses()
    
Person1 = PersonAccount("Sherif", "Bello", [5000000, 200000, 3005000], [150000, 500000], ["House", "Car"])
print(Person1.account_info())
print(f"Account balance: {Person1.account_balance()}")
print(f"Adding income of 300000.")
Person1.add_income(300000)
print(f"New account balance: {Person1.account_balance()}")
print(f"Adding expense of 100000.")
Person1.add_expense(100000)
print(f"New account balance: {Person1.account_balance()}")
print(f"Updated account info: {Person1.account_info()}")
print(f"Properties owned: {', '.join(Person1.properties)}")
print(f"Total properties: {len(Person1.properties)}")
print(f"Total income: {Person1.total_income()}")
print(f"Total expenses: {Person1.total_expenses()}")
print(f"Final account balance: {Person1.account_balance()}")
print(f"Final account info: {Person1.account_info()}")
```
