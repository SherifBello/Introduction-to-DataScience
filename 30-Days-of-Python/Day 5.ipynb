```python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
ages.sort()
print(ages)
print(f"Minimum age: {ages[0]}")
print(f"Maximum age: {ages[-1]}")
print(f"Median age: {ages[len(ages)//2]}")
print(f"Average age: {sum(ages)/len(ages)}")
print(f"Range of ages: {ages[-1] - ages[0]}")
print(f"Difference between min age and average age: {abs(ages[0] - sum(ages)/len(ages))}")
print(f"Difference between max age and average age: {abs(ages[-1] - sum(ages)/len(ages))}")
```

```python
countries_list = ['China', 'Russia','USA', 'Finland', 'Sweden', 'Norway', 'Denmark']
print(countries_list)
print(f"Number of countries in the list: {len(countries_list)}")
```

```python
n = len(countries_list)
mid = (n + 1) // 2
first_half = countries_list[:mid]
second_half = countries_list[mid:]
print("First half:", first_half)
print("Second half:", second_half)
```
