### LEVEL 1

```python
import random
import string

def random_user_id():
    chars = string.ascii_letters + string.digits
    return ''.join(random.choice(chars) for _ in range(6))
print(random_user_id())
```

```python
def user_id_gen_by_user():
    chars = string.ascii_letters + string.digits
    num_chars = int(input("Enter the number of characters: "))
    num_ids = int(input("Enter the number of IDs: "))
    ids = []

    for _ in range(num_ids):
        ids.append(''.join(random.choice(chars) for _ in range(num_chars)))
    return ids

print(user_id_gen_by_user())
```

```python
def rgb_color_gen():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return f"rgb({r}, {g}, {b})"
print(rgb_color_gen())
```

### LEVEL 2

```python
def list_of_hexa_colors():
    colors = []
    for _ in range(3):
        color = '#'
        for _ in range(6):
            color += random.choice('0123456789abcdef')
        colors.append(color)
    return colors
print(list_of_hexa_colors())
```

### LEVEL 3

```python
def shuffle_list(lst):
    random.shuffle(lst)
    return lst
print(shuffle_list([1, 2, 3, 4, 5]))
```

```python
def array_of_seven_random_numbers():
    return [random.randint(0, 9) for _ in range(7)]
    
print(array_of_seven_random_numbers())
```
