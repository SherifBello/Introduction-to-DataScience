### LEVEL 1

```python
import re
```

```python
    paragraph = 'I love teaching. If you do not love teaching what else can you love. I love Python if you do not love something which can give you all the capabilities to develop an application what else can you love.'
```

```python
frequent_words = {}
words = re.findall(r'\w+', paragraph)
for word in words:
    frequent_words[word] = frequent_words.get(word, 0) + 1
frequent_words = sorted(frequent_words.items(), key=lambda x: x[1], reverse=True)
frequent_words = [(count, word) for word, count in frequent_words]
frequent_words
```

```python
text = "The position of some particles on the horizontal x-axis are -12, -4, -3 and -1 in the negative direction, 0 at origin, 4 and 8 in the positive direction."
points = re.findall(r'-?\d+', text)
sorted_points = sorted([int(p) for p in points])
distance = sorted_points[-1] - sorted_points[0]
print("points =", points)
print("sorted_points =", sorted_points)
print("distance =", distance)
```

### LEVEL 2

```python
import keyword

def is_valid_variable(name):
    pattern = r'^[a-zA-Z_][a-zA-Z0-9_]*$'
    return bool(re.match(pattern, name)) and not keyword.iskeyword(name)
variable_names = ['first_name', 'first-name', '1first_name', 'firstname']
for name in variable_names:
    print(f"{name}: {is_valid_variable(name)}")
```

```python
sentence = '''%I $am@% a %tea@cher%, &and& I lo%#ve %tea@ching%;. There $is nothing; &as& mo@re rewarding as educa@ting &and& @emp%o@wering peo@ple. ;I found tea@ching m%o@re interesting tha@n any other %jo@bs. %Do@es thi%s mo@tivate yo@u to be a tea@cher!?'''

cleaned_text = re.sub(r'[^\w\s]', '', sentence)
print(cleaned_text)

words = re.findall(r'\w+', cleaned_text)
most_frequent_words = {}
for word in words:
    most_frequent_words[word] = most_frequent_words.get(word, 0) + 1
most_frequent_words = sorted(most_frequent_words.items(), key=lambda x: x[1], reverse=True)[:3]
print(most_frequent_words)
```
