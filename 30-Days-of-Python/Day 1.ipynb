```python
# Examples of different data types

# Number: Integer
integer_example = 42
print("Integer:", integer_example, type(integer_example))

# Number: Float
float_example = 3.14159
print("Float:", float_example, type(float_example))

# Number: Complex
complex_example = 2 + 3j
print("Complex:", complex_example, type(complex_example))

# String
string_example = "Hello, World!"
print("String:", string_example, type(string_example))

# Boolean
boolean_example = True
print("Boolean:", boolean_example, type(boolean_example))

# List
list_example = [1, 2, 3, "apple", 4.5]
print("List:", list_example, type(list_example))

# Tuple
tuple_example = (1, 2, 3, "banana", 4.5)
print("Tuple:", tuple_example, type(tuple_example))

# Set
set_example = {1, 2, 3, "cherry", 4.5}
print("Set:", set_example, type(set_example))

# Dictionary
dict_example = {"name": "Alice", "age": 30, "city": "New York"}
print("Dictionary:", dict_example, type(dict_example))
```

```python
import math

# Points
point1 = (2, 3)
point2 = (10, 8)

# Euclidean distance
distance = math.sqrt((point2[0] - point1[0])**2 + (point2[1] - point1[1])**2)
print(distance)
```
