Question 4

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]
flattened_countries = [
    [country.upper(), country.upper()[:3], city.upper()]
    for row in countries
    for country, city in row
]
print(flattened_countries)
```

Question 5

```python
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]
countries_dicts = [
    {'country': country.upper(), 'city': city.upper()}
    for row in countries
    for country, city in row
]
countries_dicts
```

Question 6

```python
names = [[('Asabeneh', 'Yetayeh')], [('David', 'Smith')], [('Donald', 'Trump')], [('Bill', 'Gates')]]
concatenated_names = [f"{first} {last}" for row in names for first, last in row]
print(concatenated_names)
```

Question 7

```python
slope = lambda x1, y1, x2, y2: (y2 - y1) / (x2 - x1)
y_intercept = lambda x, y, m: y - m * x

# example
m = slope(1, 2, 3, 8)
b = y_intercept(1, 2, m)
print(m, b)
```
