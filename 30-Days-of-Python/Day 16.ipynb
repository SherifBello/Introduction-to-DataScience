```python
import json
from collections import Counter
from typing import List, Tuple

def most_spoken_languages(path: str = 'data/countries_data.json', top_n: int = 10) -> List[Tuple[int, str]]:
    with open(path, encoding='utf-8') as f:
        countries = json.load(f)
    counter = Counter()
    for country in countries:
        langs = country.get('languages') or country.get('language')
        if not langs:
            continue
        if isinstance(langs, str):
            counter[langs] += 1
        else:
            for l in langs:
                if isinstance(l, str):
                    counter[l] += 1
                elif isinstance(l, dict):
                    name = l.get('name') or l.get('language')
                    if name:
                        counter[name] += 1
    return counter.most_common(top_n)

most_spoken_languages()
```

```python
def most_populated_countries(filename='./data/countries_data.json', top_n=10):
    with open(filename, encoding='utf-8') as f:
        countries = json.load(f)
    sorted_countries = sorted(countries, key=lambda x: x['population'], reverse=True)
    top = sorted_countries[:top_n]
    return [{'country': c['name'], 'population': c['population']} for c in top]
most_populated_countries()
```

```python
import re

def find_most_frequent_words(filename, top_n=10):
    with open(filename, 'r', encoding='utf-8') as fh:
        text = fh.read().lower()
    words = re.findall(r"[a-z]+", text)
    return Counter(words).most_common(top_n)

for label, filename in [
    ("Obama", "data/obama_speech.txt"),
    ("Michelle", "data/michelle_obama_speech.txt"),
    ("Trump", "data/donald_speech.txt"),
    ("Melina", "data/melina_trump_speech.txt"),
]:
    print(label, find_most_frequent_words(filename))
```

```python
from pathlib import Path
import math

def load_text(source):
    if isinstance(source, str):
        path = Path(source)
        if path.exists() and path.is_file():
            return path.read_text(encoding='utf-8')
        return source
    raise TypeError("Source must be a file path or a string.")

def load_stop_words(data_dir='data'):
    data_path = Path(data_dir)
    stop_files = sorted(data_path.glob('*stop*'))
    stop_words = set()
    for path in stop_files:
        text = path.read_text(encoding='utf-8')
        stop_words.update(re.findall(r"[a-z']+", text.lower()))
    if not stop_words:
        # fallback minimal set
        stop_words = {
            'a', 'an', 'and', 'the', 'is', 'in', 'it', 'of', 'for', 'on',
            'to', 'with', 'that', 'this', 'as', 'at', 'by', 'from', 'be',
            'are', 'was', 'were', 'or', 'not', 'but', 'if', 'so', 'we', 'you',
            'i', 'he', 'she', 'they', 'them', 'their', 'will', 'has', 'have',
            'had', 'our', 'us', 'been', 'which'
        }
    return stop_words

def clean_text(text):
    text = load_text(text)
    text = text.lower()
    text = re.sub(r"[^a-z0-9'\s]", " ", text)
    return re.findall(r"[a-z']+", text)

def remove_support_words(words, stop_words):
    return [w for w in words if w and w not in stop_words]

def check_text_similarity(text1, text2, stop_words=None):
    stop_words = stop_words or load_stop_words()
    tokens1 = remove_support_words(clean_text(text1), stop_words)
    tokens2 = remove_support_words(clean_text(text2), stop_words)

    if not tokens1 or not tokens2:
        return {
            'cosine_similarity': 0.0,
            'jaccard_similarity': 0.0,
            'common_words': [],
            'tokens_text1': len(tokens1),
            'tokens_text2': len(tokens2)
        }

    vec1 = Counter(tokens1)
    vec2 = Counter(tokens2)

    dot = sum(vec1[word] * vec2[word] for word in vec1 if word in vec2)
    norm1 = math.sqrt(sum(count * count for count in vec1.values()))
    norm2 = math.sqrt(sum(count * count for count in vec2.values()))
    cosine = dot / (norm1 * norm2) if norm1 and norm2 else 0.0

    set1 = set(vec1)
    set2 = set(vec2)
    jaccard = len(set1 & set2) / len(set1 | set2) if set1 or set2 else 0.0

    common_words = sorted(
        [(word, vec1[word], vec2[word]) for word in set1 & set2],
        key=lambda item: -(item[1] + item[2])
    )[:20]

    return {
        'cosine_similarity': round(cosine, 4),
        'jaccard_similarity': round(jaccard, 4),
        'common_words': common_words,
        'tokens_text1': len(tokens1),
        'tokens_text2': len(tokens2),
    }

similarity_results = check_text_similarity(
    'data/michelle_obama_speech.txt',
    'data/melina_trump_speech.txt'
)

print("Similarity between Michelle's and Melina's speeches:")
print(f"  Cosine similarity: {similarity_results['cosine_similarity']}")
print(f"  Jaccard similarity: {similarity_results['jaccard_similarity']}")
print(f"  Tokens in Michelle speech: {similarity_results['tokens_text1']}")
print(f"  Tokens in Melina speech: {similarity_results['tokens_text2']}")
print("  Top common words:", similarity_results['common_words'][:10])
```

```python
from pathlib import Path
import re

csv_path = Path('data/hacker_news.csv')
with csv_path.open('r', encoding='utf-8', errors='ignore') as fh:
    lines = fh.readlines()

python_count = 0
javascript_count = 0
java_count = 0

for line in lines:
    lower_line = line.lower()
    if 'python' in lower_line:
        python_count += 1
    if 'javascript' in lower_line:
        javascript_count += 1
    if re.search(r'\bjava\b', line, re.IGNORECASE) and not re.search(r'\bjavascript\b', line, re.IGNORECASE):
        java_count += 1

print("Lines containing python or Python:", python_count)
print("Lines containing JavaScript/javascript/Javascript:", javascript_count)
print("Lines containing Java but not JavaScript:", java_count)
```
