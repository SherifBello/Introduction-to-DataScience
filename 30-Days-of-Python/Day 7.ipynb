### LEVEL THREE

```python
print(age)

len(age)
```

```python
age_set = set(age)
print(age_set)
print(len(age_set))
```

```python
age_set2 = set(age)  # convert list to set (duplicates removed)
len_list = len(age)
len_set = len(age_set2)

print("age list:", age)
print("age set:", age_set2)
print("length list:", len_list)
print("length set:", len_set)
print("bigger:", "list" if len_list > len_set else "set" if len_set > len_list else "equal")
```

A String: is any data type written as a text in quotation

A List: is a collection which is ordered and changeable(Mutable)

A Tuble: is a collection which is unordered and unchangeable(Immutable)

A Set: is a collection which is unorderd, unmmodifiable and un-indexed.
