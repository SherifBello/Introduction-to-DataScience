```python
import math

# Calculate slope, x-intercept, and y-intercept of y = 2x - 2
slope_line = 2
y_intercept = -2
x_intercept = 1  # From 0 = 2x - 2 => x = 1

print('8.'f"Slope: {slope_line}")
print('  'f"X-intercept: {x_intercept}")
print('  'f"Y-intercept: {y_intercept}"'\n')

# Calculate slope and Euclidean distance between (2,2) and (6,10)

x1, y1 = 2, 2
x2, y2 = 6, 10

slope_points = (y2 - y1) / (x2 - x1)
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print('9.'f"Slope between points: {slope_points}")
print('  'f"Euclidean distance: {distance}")
```

```python
# Define the quadratic function
def calculate_y(x):
    return x**2 + 6*x + 9

# Try different x values and find when y is 0
for x in range(-5, 5):
    y = calculate_y(x)
    print(f"For x = {x}, y = {y}")
    if y == 0:
        print(f"y is 0 when x = {x}")
        break
```
